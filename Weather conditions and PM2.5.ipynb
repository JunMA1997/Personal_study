## (4) our work:

### (i) goal:

* 初始目标：$PM_{2.5}$变化率与环境条件之间的联系
* * 通过大量的数据分析以及部分尝试发现，变化率与环境条件之间的联系较为复杂，通过简单的数据变换并不能归纳总结出较为精确的数据模型。
* 改变后目标： $PM_{2.5}$浓度与环境条件之间的联系
* * 数据分析
* * * 通过将数据按时间序列绘制成二维散点折线图,发现$PM_{2.5}$浓度具有很强的季节性，在10月到4月之间的数值普遍高于其他月份的数值。通过经验判断，这是因为取暖季及冬天特有的气候条件所导致的。
* * * 通过经验，我们可以想象$PM_{2.5}$浓度在一天中的排放量应该是有迹可循的。我们推测：如果样本量足够大，那么我们应该可以在保持气候条件变化较小的的情况下预测日期相近的$PM_{2.5}$浓度。然而通过数据可视化分析，我们并没有能够找出明显的日变化曲线。
* * * 通过数据分析可以发现，$PM_{2.5}$浓度与当前时间的环境条件具有很强的相关性。在相关性分析中，该浓度尤其与其他污染物浓度具有更强的相关性。但与此同时，也与风速，露点温度以及降雨有着较强的相关性。通过查阅资料，我们总结出我们的第一个模型：$$PM_{2.5}=Emission+Difussion$$

* * RNN（LSTM）:
* * * 建立依据及目的：
* * * * 依据：
* * * * * 在较短时间内，我们将$PM_{2.5}$的排放量视为一个整体，即在这段时间内我们认为$PM_{2.5}$的排放量不变。由此，$PM_{2.5}$的排放量将仅受扩散条件即天气条件的影响。
* * * * 目的：
* * * * * 在最初的几个模型中，我们期望能获得$PM_{2.5}$浓度与天气条件之间的关系。通过EDA分析及实际尝试，线性模型很难准确预测到$PM_{2.5}$浓度：大部分预测的预测值都与实际值存在较大偏差。但是机器学习及神经网络的出现使得这个想法变成了可能。
* * * 模型概述：
* * * * 该模型使用24小时内天气条件作为输入，预测下一小时$PM_{2.5}$浓度，因为$PM_{2.5}$具有季节性变化，因此选取取暖季数据作为分析数据
* * * 结果总结：
* * * * 该模型预测出的预测集及实际结果值具有基本相似的趋势，其$R^2$得分为91%。
